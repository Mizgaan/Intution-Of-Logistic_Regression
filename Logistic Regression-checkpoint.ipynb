```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

# Formulla For Sigmoid function in logestic regression y=1/1+(e)^-mx+c


```python
x= np.arange(-50,30)
```

```python
print(len(x),'=> length')
x
```

```python
c= np.arange(-10,20) #Intercept 
act_opt=[]
pre_opt={}
for i in x:
    if(i<10):
        act_opt.append(0)
    else:
        act_opt.append(1)   
    
## Giving various range of intercept
for i in c:
    predict_opt=1/(1+np.exp(-1*(x)+i))  #Applied Sigmoid Function :- 1/1+(e)^-mx+c
    pre_opt[i]=predict_opt
    plt.plot(x,pre_opt[i],label=i,)
     #  plt.plot(x,pre_opt[i],label=i)

x_smooth=np.linspace(x.max(),x.min())             # For Smoothing the sigmoid curve of actual oputput
y_smooth=make_interp_spline(x,act_opt)(x_smooth)

plt.plot(x_smooth,y_smooth,label='actual',c='red')   
plt.legend()
plt.grid()
plt.title('Logistic Regression')
plt.ylabel('Y axis')
plt.xlabel('X axis')
#plt.show()
#print( pre_opt[i])


   # plt.plot(x,np.round( pre_opt[i],1))
```
